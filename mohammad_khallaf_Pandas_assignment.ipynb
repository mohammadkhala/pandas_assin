**correlation coefficient**



*   The correlation coefficient is an equation that is used to determine

      the strength of the relation between two variables


*   The correlation coefficient always lies between -1 to +1
*    -1 represents X and Y are negatively correlated
*   +1 represents X and Y are positively correlated



---



---


r = n (xy)  x y /  [n* (x2  (x)2)] * [n* (y2  (y)2)]


*   r = correlation coefficient

*   n = number of observations

*   x = 1st variable in the context



*   y = 2nd variable




---

Absolute Value of r Strength of Relationship


r < 0.3  ==>  None or very weak


0.3 < r <0.5   ==>  Weak


0.5 < r < 0.7  ==>   Moderate


r > 0.7  ==>   Strong

```python
# Solution 6.3
def correlation_coefficient(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x[i]*y[i] for i in range(n))
    sum_x_sq = sum(x[i]**2 for i in range(n))
    sum_y_sq = sum(y[i]**2 for i in range(n))

    numerator = (n * sum_xy) - (sum_x * sum_y)
    denominator = ((n * sum_x_sq) - (sum_x ** 2)) * ((n * sum_y_sq) - (sum_y ** 2))

    correlation = numerator / (denominator ** 0.5)
    return correlation


total_rooms = df['total_rooms']
median_house_value = df['median_house_value']
correlation = correlation_coefficient(total_rooms, median_house_value)
print("\nCorrelation coefficient between total_rooms and median_house_value:", correlation)
filtered_df = df[(df['population'] > 5000) & (df['median_income'] > 7)]
query_result = filtered_df[['median_house_value', 'median_income']]
print("\nQuery result:")
query_result
```
